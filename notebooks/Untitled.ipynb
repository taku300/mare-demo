```python
import os
from langchain_openai import AzureOpenAI
from langchain_core.messages import HumanMessage
from langchain_core.runnables.history import RunnableWithMessageHistory
from langchain_community.chat_message_histories import SQLChatMessageHistory

# Azure OpenAI APIキーとエンドポイントの設定
os.environ["OPENAI_API_VERSION"] = "2023-12-01-preview"
os.environ["AZURE_OPENAI_ENDPOINT"] = "https://hon302-openai.openai.azure.com/"
os.environ["AZURE_OPENAI_API_KEY"] = "1ccc2b4e714e473f925a763bd42abcec"

# Azure OpenAIモデルの設定
llm = AzureOpenAI(
    deployment_name="gpt-35-turbo-instruct-0914",
)

# セッション履歴を管理する関数
def get_session_history(session_id):
    return SQLChatMessageHistory(session_id, "sqlite:///memory.db")

# RunnableWithMessageHistoryの設定
runnable_with_history = RunnableWithMessageHistory(
    llm,
    get_session_history,
)

# あらかじめ決まった質問リスト
questions = [
    "こんにちは、今日はどんな天気ですか？",
    "AIについてどう思いますか？",
    "最近のニュースについて教えてください。",
    "あなたはどんな能力がありますか？",
    "将来の展望についてどう思いますか？"
]

# 対話の実行
def run_conversation(session_id, question):
    response = runnable_with_history.invoke(
        [HumanMessage(content=question)],
        config={"configurable": {"session_id": session_id}},
    )
    return response

# 質問を順番に実行
session_id = "conversation_1"
for question in questions:
    print(f"あなた: {question}")

    # 対話の実行と応答の表示
    ai_response = run_conversation(session_id, question)
    print(f"AI: {ai_response.content}\n")
```
